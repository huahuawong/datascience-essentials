## Naive Bayes
### It is a classification method that is based on the Bayes theorem. It is simple and yet can be effective in performing classification. The Bayes theorem is as follow:



### We should bear in mind that in naive bayes, the features are assumed to be independent of one another. This assumption simplifies computation, and hence it is considered as naive. This assumption is called class conditional independence.

### How does the naive bayes classifier work?
1. Calculate the prior probability for each class labels
2. Find the likelihood with each features for each class
3. Use Bayes theorem to find the posterior probability
4. Observe which class has a higher probability, given the input belongs to the higher probability class.

### Wait, what's the difference between prior, likelihood and posterior?


As we know, posterior = likelihood * prior

1. Likelihood is the probability of falling under a specific category or class. It can be viewed as a function of the parameters. One of the ways that we can use this for ML model, is MLE, Maximum likelihood estimation, where we maximize the parameter values according to maximum likelihood. We take the likelihood function, and attempt to maximize it by changing the parameter settings (keeping the observed data constant).

2. Prior is the probability distribution representing knowledge or uncertainty of a data object prior or before observing it
3. Posterior is the conditional probability distribution representing what parameters are likely after observing the data object

If we refer to the formula above, P(A|B) would be the posteior, P(B|A) would be the likelihood. P(A) and P(B) would be the priors for the predictor and class respectively.
